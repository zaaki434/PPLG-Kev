#Permainan TIC TAC TOE menggunakan Python

```python
# definisikan papan permainan berupa kotak 3x3
def papan_permainan(papan):
    print(papan[0] + " | " + papan[1] + " | " + papan[2])
    print("-" * 9)
    print(papan[3] + " | " + papan[4] + " | " + papan[5])
    print("-" * 9)
    print(papan[6] + " | " + papan[7] + " | " + papan[8])

#fungsi untuk pemenang
def cek_pemenang(papan, mark):
    # Check barisan
    for i in range(0, 9, 3):
        if papan[i] == papan[i+1] == papan[i+2] == mark:
            return True
    # Check kolom
    for i in range(0, 3):
        if papan[i] == papan[i+3] == papan[i+6] == mark:
            return True
    # Check diagonal
    if papan[0] == papan[4] == papan[8] == mark or papan[2] == papan[4] == papan[6] == mark:
        return True
    # No winner
    return False

# Define a function to get the pemain input and validate it
def get_input(papan, mark):
    while True:
        try:
            posisi = int(input(f"pemain {mark}, pilih kotak 1 sampai 9: ")) - 1
            if posisi not in range(0, 9):
                print("input invalid, silahkan pilih ulang.")
            elif papan[posisi] != " ":
                print("Kotak sudah di pilih, silahkan pilih kotak yang lain .")
            else:
                return posisi
        except ValueError:
            print("Input invalid, silahkan masukkan angka kotak .")

# Define a function to switch the turn between the pemain
def switch_turn(mark):
    if mark == "X":
        return "O"
    else:
        return "X"

# fungsi untuk menjalankan perulangan game
def play_game():
    # load papan dan mark/tanda
    papan = [" "] * 9
    mark = "X"

    # Tampilkan papan
    papan_permainan(papan)

    # ulangi/loop sampai ada yang menang atau seri
    while True:
        # tampilkan papan yang sudah di input
        posisi = get_input(papan, mark)
        papan[posisi] = mark

        # tampilkan papan yang di perbarui
        papan_permainan(papan)

        # cek apakah ada pemenang atau draw
        if cek_pemenang(papan, mark):
            print(f"pemain {mark} menang")
            break
        elif " " not in papan:
            print("Draw")
            break

        # ganti giliran pemain
        mark = switch_turn(mark)

# mulai permainan
play_game()
```

- Tentukan fungsi untuk menampilkan papan
- Tentukan fungsi untuk memeriksa apakah ada pemenang
- Tentukan fungsi untuk mendapatkan masukan pemain dan memvalidasinya
- Tentukan fungsi untuk mengalihkan giliran antar pemain
- Tentukan fungsi untuk menjalankan loop utama permainan
