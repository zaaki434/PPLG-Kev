Operasi arimatika pengurangan

```
a=5
b=2
c=a-b #value dari variabel C adalah pengurangan dari value variabel a & b
print(c)
```

Operasi arimatika perkalian

```
a=8
b=5
c=a*b #value dari variabel C adalah perkalian dari value variabel a & b
print(c)
```

Operasi arimatika pembagian

```
a=10
b=5
c=a/b #value dari variabel C adalah pembagian dari value variabel a & b
print(c)
```

Operasi arimatika Hasil

Operasi arimatika akar

```
a=30
b=5
c=a//b #value dari variabel c adalah akar dari 30
print(c)
```

Operasi arimatika pangkat

```
a=10
b=5
c=a**b # value dari variabel c adalah isi dari variabel a yang di pangkatkan variabel b
print(c)
```

Operasi arimatika persamaan

```
a=10
b=10
c=a==b #a dengan b memiliki value yang sama
print(c)
```

```
a=10
b=11
c=a!=b
print(c) #a dengan b berbeda tetapi hasilnya akan di paksa untuk true
```
