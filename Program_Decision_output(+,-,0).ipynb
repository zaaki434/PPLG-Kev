#Buatlah sebuah program untuk menentukan bilangan positif,negatif dan bilangan 0

```python
#Menginput Angka
angka = float(input("Tulis Sebuah Angka: "))

#jika angka lebi dari 0 maka akan print positif
if angka > 0:
  print("Angka Positif")

#jika value 0 maka akan print 0
elif angka == 0:
  print("Angka Nol")

#jika value kurang dari 0 maka akan print negatif
else:
  print("AngkaÂ Negatif")
```

- input data
- jika data yang di input value nya lebih dari 0 maka output nya positif
- jika data yang di input value nya kurang dari 0 maka output nya negatif
- jika data yang di input value nya 0 maka output nya nol atau netral
