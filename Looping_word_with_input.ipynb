#Buatlah sebuah program sederhana menggunakan perulangan dan algoritma dan flowchartnya

Program perulangan sederhana menggunakann "for"

```python
#Menginput Angka
kalimat1 = str(input("masukkan kata yg akan di print:  "))
angka = int(input("Jumlah yang akan di print:  "))

for angka in range(angka):
  print (kalimat1)  #akan mengeprint kalimat sesuai dengan kalimat yang di input
else: #akan mengeprint selesai jika perulangan sudah memenuhi syarat
  print ("selesai")
```

- masukkan kalimat
- masukkan value berupa angka untuk berapa banyak kalimat akan di looping
- setelah looping berhasil akan mengeprint kalimat selesai
- hasil akan di tampilkan
